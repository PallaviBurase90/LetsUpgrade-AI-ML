Q.5 Write a program to convert fahrenheit to kelvin and celsius both.

```
#fahrenheit to kelvin and celsius
F=52 # Temprature in Fahrenheit 
celsius=((F - 32) / 1.8)
print("Fahrenheit in Celsius is = ", celsius)
kelvin = 273.15 + celsius
print("Fahrenheit in Kelvin is = ", kelvin)
```
