Q.1 Research on whether addition, subtraction, multiplication, division, floor division and modulo
operations be performed on complex numbers. Based on your study, implement a Python
program to demonstrate these operations.

```
a=complex(input("Enter complex number of form a+bj: "))
b=complex(input("Enter complex number of form a+bj: "))

print("Addition of two complex numbers : ",a+b) #Addition

print("Subtraction of two complex numbers : ",a-b) #Substraction

print("Multiplication of two complex numbers : ",a*b) # Multiplication

print("Division of two complex numbers : ",a/b)  #division

print("The modulus of ",a," is", abs(a)) #modulo 
print("The modulus of ",b," is", abs(b))  #modulo

print("Floor divisiom", a//b)  #Floor division Cant perform on complex number
```
