Q.1 Write a Python program to find the first 20 non-even prime natural numbers

```
from sympy  import isprime
count=0
for i in range(100):  
   if (isprime(i) and count<20 ):
       if i%2 !=0:
           print(i)
           count+=1
    
```

Q.3 Write a Python program to check if the given string is a Palindrome or Anagram or None of them.
Display the message accordingly to the user.

```
def palin():
    a=list(input("Enter your String"))
    print(a)
    for i in range(0,len(a)):
        if a[i]==a[len(a)-(i+1)]:
            palindrone=True
        else:
            palindrone=False
            break
    if palindrone == True:
        print("Given String is Palindrone")
    else:
        print("Given String is not palindrone")
    
def anagram():
    first_string=list(input("Enter your First string"))
    print(first_string)
    second_string=list(input("Enter Your Second String"))
    print(second_string)
    for i in second_string:
        if i in first_string:
            Anagram_condition=True
        else:
            Anagram_condition=False
            break
        
    if Anagram_condition==True:
        print("Given string is Anagram")
    else:
        print("Given string is not an Anagram")
        
Choice=int(input("Select 1 To Check If String Is Palindrone Or Not /n Select 2 To Check If String Is Anagram Or Not"))

Anagram_condition=False
palindrone=False

if Choice==1:
    palin()
else:
    anagram()
    
    
if palindrone==True and Anagram_condition ==True:
    print("Given String is palindrone & Anagram")
elif palindrone==True:
    print("Given String is palidrone")
elif Anagram_condition==True:
    print("Given String is Anagram")
else:
    print("None Of them!")
```

Q.4 Write a Python's user defined function that removes all the additional characters from the string
and converts it finally to lower case using built-in lower(). eg: If the string is "Dr. Darshan Ingle
@AI-ML Trainer", then the output be "drdarshaningleaimltrainer".

```
def remove_str(string):
    alpha= "abcdefghijklmnopqrstuvwxyz1234567890"
    string=string.lower().replace(" ", " ")
    new_str=""
    for i in string:
        if i in alpha:
           new_str=new_str +i
    print(new_str)
remove_str("Pallavi@.--- Burase")    
```
